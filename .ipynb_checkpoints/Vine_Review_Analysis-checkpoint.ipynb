# Deliverable 2 - Determine Bias of Vine Reviews

```python
import pandas as pd

# File to Load
vine_file_to_load = "vine_table.csv"

# read vine table data
vine_df = pd.read_csv(vine_file_to_load)
vine_df
```

```python
# Filter the data and create a new DataFrame or table to retrieve 
# all the rows where the total_votes count is equal to or greater than 20
helpful_votes_df = vine_df[vine_df["total_votes"] >= 20]
helpful_votes_df
```

```python
# Filter the new DataFrame created in Step 1 and create a new DataFrame to retrieve all the rows where 
# the number of helpful_votes divided by total_votes is equal to or greater than 50%
good_helpful_df = helpful_votes_df[helpful_votes_df["helpful_votes"] / helpful_votes_df["total_votes"] >= 0.5]
good_helpful_df
```

```python
vine_paid_df = good_helpful_df[good_helpful_df.vine == "Y"]
vine_paid_df
```

```python
vine_unpaid_df = good_helpful_df[good_helpful_df.vine == "N"]
vine_unpaid_df
```

```python
# total number of reviews
len(vine_df)
```

```python
# total number of 5 star reviews
len(vine_df[vine_df["star_rating"] == 5])
```

```python
# total number of paid 5 start reviews
len(vine_paid_df[vine_paid_df["star_rating"] == 5])
```

```python
# total number of unpaid 5 star reviews
len(vine_unpaid_df[vine_unpaid_df["star_rating"] == 5])
```

```python
# percentage of paid 5 star reviews
len(vine_paid_df[vine_paid_df["star_rating"] == 5]) / len(vine_df) * 100
```

```python
# percentage of unpaid 5 star reviews
len(vine_unpaid_df[vine_unpaid_df["star_rating"] == 5]) / len(vine_df) * 100
```
